```python
from sklearn.datasets import make_blobs
from scipy.stats import norm
from numpy import mean,std
```

```python
X,y=make_blobs(n_samples=100,centers=2,n_features=2, random_state=1)
```

```python
X
```

```python
y
```

```python
# calculate priors 

Xy0=X[y==0]
Xy0
```

```python
Xy1=X[y==1]
Xy1
```

```python
print(Xy0.shape,Xy1.shape)
```

```python
priory0=len(Xy0)/len(X)
priory1=len(Xy1)/len(X)
```

```python
print(priory0,priory1)
```

```python
def fit_distribution(data):
    mu=mean(data)
    sigma=std(data)
    
    #print(mu,sigma)
    dist=norm(mu,sigma)
    return dist
    
```

```python
fit_distribution(Xy1[:,0])
```

```python
X1y0=fit_distribution(Xy0[:,0])
X2y0=fit_distribution(Xy0[:,1])
X1y1=fit_distribution(Xy1[:,0])
X2y1=fit_distribution(Xy1[:,1])
```

```python
def prob(X,prior,dist1,dist2):
    return prior*dist1.pdf(X[0])*dist2.pdf(X[1])
```

```python
#classify one example
Xsample,ysample=X[0],y[0]
print(Xsample,ysample)
```

```python
py0=prob(Xsample,priory0,X1y0,X2y0)
py1=prob(Xsample,priory1,X1y1,X2y1)
print(py0,py1)
```

```python
if py1>py0:
    print(1)
else:
    print(0)
```
